### Option 1: Use Pandas to create the contacts DataFrame.

```python
# Iterate through the contact_info_df and convert each row to a dictionary.
import json

dict_list = []

for index, row in contact_info_df.iterrows():
    dictionary = row.to_dict()
    dict_list.append(dictionary)
 
values = []
column_names = []
    
for dictionary in dict_list:
    contact_info = dictionary['contact_info']
    contact_data = json.loads(contact_info)
    
    columns = [k for k, v in contact_data.items()]
    
    row_values = [v for k,v in contact_data.items()]
    
    column_names.append(columns)
    values.append(row_values)
    
    contact_id = contact_data['contact_id']
    name = contact_data['name']
    email = contact_data['email']

new_temp_contact_df = pd.DataFrame(values, columns=column_names[0])
new_temp_contact_df.head()


    
```

```python
new_temp_contact_df = new_temp_contact_df.rename(columns={"contact_id":"Contact ID", "name": "Name", "email": "Email"})
new_temp_contact_df
```

```python
# Check the datatypes.
new_temp_contact_df.dtypes
```

```python
# Create a "first"name" and "last_name" column with the first and last names from the "name" column. 

new_temp_contact_df[['First Name','Last Name']] = new_temp_contact_df['Name'].str.split(' ', 1, expand=True)

new_temp_contact_df = new_temp_contact_df.drop('Name', axis = 1)

new_temp_contact_df
# Drop the contact_name column

```

```python
# Reorder the columns
new_temp_contact_df = new_temp_contact_df[['Contact ID', 'First Name', 'Last Name', 'Email']]
new_temp_contact_df
```
